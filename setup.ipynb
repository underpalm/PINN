# Mathematisches Verständnis: Die Ableitung

## 1. Das "2-zu-1"-Verhältnis
Wenn die Ableitung an einem Punkt $x$ den Wert $2$ hat, also 
$$f'(x) = 2$$
dann flüstert dir die Kurve an dieser Stelle folgendes Gesetz zu:

> "Wenn du mich jetzt hier um einen winzigen Schritt nach rechts schubst ($+1$), werde ich mich doppelt so stark nach oben bewegen ($+2$)."

**Die Ableitung ist also ein Multiplikator.** Es beschreibt die **Sensitivität** der Funktion: Wie stark reagiert der Output ($y$) auf eine winzige Änderung des Inputs ($x$)?

# Kurz gesagt: Die zweite Ableitung verrät dir, wie sich die Steigung selbst verändert.

Wenn die erste Ableitung der Tacho (die Geschwindigkeit) ist, dann ist die zweite Ableitung das Gaspedal oder die Bremse (die Beschleunigung).
